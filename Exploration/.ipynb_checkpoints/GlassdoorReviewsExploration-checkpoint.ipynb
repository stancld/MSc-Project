### **Number of reviews (+average rating) per company and month**

<hr>

Reviews with date of (0, 0) are those ones which are labelled as featured reviews (they usually posses of a large number of helpfulness likes) and thus a date is unavailable for them. As such, it will be the best to drop them.

```python
def reviewsPerCompanyAndMonth(companies):
    fig, ax = plt.subplots(nrows=len(companies), ncols=1, figsize=(10,5*len(companies)))
    
    for i, company in enumerate(companies):
        (data[data.Company==company]
         .groupby(['Company', 'Year', 'Month'])
         .count()
         .ReviewTitle
         .plot.bar(ax=ax[i])
         .set_xlabel(company)
        )

    plt.tight_layout()
```

```python
companies = data.Company.unique()

reviewsPerCompanyAndMonth(companies)
```

```python
def ratingsPerCompanyAndMonth(companies):
    fig, ax = plt.subplots(nrows=len(companies), ncols=1, figsize=(10,5*len(companies)))
    
    for i, company in enumerate(companies):
        (data[data.Company==company]
         .groupby(['Company', 'Year', 'Month'])
         .Rating
         .mean()
         .plot
         .bar(ax=ax[i], ylim=(0,5))
         .set_xlabel(company)
        )

    plt.tight_layout()
```

```python
companies = data.Company.unique()

ratingsPerCompanyAndMonth(companies)
```

### Some basic statistics

```python
def firstDecile(x):
    return x.quantile(0.1)

def firstQuartile(x):
    return x.quantile(0.25)

def median(x):
    return x.quantile(0.5)

def thirdQuartile(x):
    return x.quantile(0.75)

def ninthDecile(x):
    return x.quantile(0.9)
```

```python
(data
 .groupby('Company')
 .agg({'Rating': ['mean', 'std', 'skew', firstDecile, firstQuartile, median, thirdQuartile, ninthDecile]})
)
```

```python
def reviewsHistogram(companies):
    fig, ax = plt.subplots(nrows=1, ncols=len(companies), figsize=(20,4))
    
    for i, company in enumerate(companies):
        (data[data.Company==company]
         .groupby('Rating')
         .count()
         .ReviewTitle
         .plot.bar(ax=ax[i])
         .set_xlabel(company)
        )

    plt.tight_layout()
```

```python
companies = data.Company.unique()

reviewsHistogram(companies)
```

### **Portion (+ mean, std and skew stats) of reviews written by past nad current employees**

```python
data['Current|Former'] = data.JobTitle.apply(lambda x: x.split(' - ')[0])
```

```python
pd.set_option('display.max_rows', None)

(data[data['Current|Former'].isin(['Current Employee', 'Former Employee'])]
 .groupby(['Company', 'Current|Former'])
 .Rating
 .agg(['count', 'mean', 'std', 'skew'])
)
```

```python
(data[data['Current|Former'].isin(['Current Employee', 'Former Employee'])]
 .groupby(['Company', 'Current|Former'])
 .Rating
 .mean()
 .unstack(fill_value=0)
 .plot
 .bar()
)
```

```python
(data[data['Current|Former'].isin(['Current Employee', 'Former Employee'])]
 .groupby(['Company', 'Current|Former'])
 .Rating
 .std()
 .unstack(fill_value=0)
 .plot
 .bar()
)
```

```python
(data[data['Current|Former'].isin(['Current Employee', 'Former Employee'])]
 .groupby(['Company', 'Current|Former'])
 .Rating
 .skew()
 .unstack(fill_value=0)
 .plot
 .bar()
)
```
